```python
import numpy as np
# Create a sample dataset
df = df_with_0
# df = egf
# Estimate the mean and variance for each column
means = df.mean()
variances = df.var()

# Simulate Gaussian data based on the estimates
simulated_data = pd.DataFrame()
for column in df.columns:
    mean = means[column]
    std_dev = np.sqrt(variances[column])
    simulated_data[column] = np.random.normal(loc=mean, scale=std_dev, size=len(df))

# Print estimated means and variances
print("Estimated Means:")
print(means)
print("\nEstimated Variances:")
print(variances)

# Print the first few rows of the simulated data
print("\nSimulated Data:")
print(simulated_data.head())
```
